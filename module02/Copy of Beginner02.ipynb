```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000

# Define bins and labels
bins = [0, 10000, 20000, float('inf')]
labels = ['low', 'medium', 'high']

# Use cut to segment and sort the data values into bins
# mapping the company_price on the 'CarCompany' column in the dataset
CarData['CarCompany'] = pd.cut(CarData['price'], bins=bins, labels=labels, right=False)



...

# checking the column's unique values

#CarData.CarCompany.unique()

CarData.CarCompany.unique()
```

```python
columns_to_drop = ['car_ID', 'CarName']
CarData.drop(columns=columns_to_drop, inplace=True)
```

```python
#creating dummy variable for catwegorical variables
CarData = pd.get_dummies(CarData, dtype =int)

#checking the dataset
CarData.head()
```

```python
CarData.shape
```

```python
# importing required library
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd


# We specify this so that the train and test data set always have the same rows, respectively
X = CarData.drop(columns=['price'])
y = CarData['price']


# We specify this so that the train and test data set always have the same rows, respectively
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train model
model = LinearRegression()
train = model.fit(X_train_scaled, y_train)

# Now, you can use the trained model for predictions or further evaluation
```

```python
# Training
predicted_price = train.predict(X_test_scaled)
```

```python
# Show predictions
predicted_price
```

```python
# Plotting y_test and y_pred to understand the spread.

import matplotlib.pyplot as plt

plt.scatter(y_test, predicted_price, color='blue', alpha=0.5)  # Alpha adjusts the transparency of points
plt.xlabel('Actual Price (y_test)')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_true=y_test, y_pred=predicted_price))
```
